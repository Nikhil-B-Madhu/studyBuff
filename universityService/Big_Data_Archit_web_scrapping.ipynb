## Get Course Names from Course Catalog

```python
i = 0
j = 0
for i in range(1, link_pd.shape[0]+1):
    url = urllib.request.urlopen(link_pd['full_links'][i])
    content = url.read()
    soup = bs(content, 'lxml')
    table = soup.findAll('div',attrs={"class":"courseblock"})

    for x in table:
        print(x.find('p').text)
        j+=1    
        get_courses_details.loc[j] = [x.find('p').text,link_pd['full_links'][i]]
```

```python
get_courses_details.head()
```

```python
courses_desc['courses_id'] = get_courses_details['Courses'].str.split(' ', n = 1, expand = True)[0].replace(u'\xa0', u' ')
courses_desc['course_name'] = get_courses_details["Courses"].str.split(")", n = 1, expand = True)[1].replace(u'\xa0', u' ')
courses_desc['department_id'] = get_courses_details["Courses"].str.split(" ", n = 1, expand = True)[0].str.split("\xa0", expand = True)[0]
```
